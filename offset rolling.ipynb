```
R=1;
r=0.8;
M=1;
m=1;
g=9.8;
alpha=30 Pi/180;
beta=0;
moi=2/5 M R^2;

a[t_]:=R theta[t];
xc[t_]:=-a[t] Cos[alpha];
yc[t_]:=a[t] Sin[alpha];
x[t_]:=xc[t]+r Cos[theta[t]+beta];
y[t_]:=yc[t]+r Sin[theta[t]+beta];

KE=1/2 moi theta'[t]^2+1/2 m (x'[t]^2+y'[t]^2);
PE=M g yc[t]+m g y[t];
L=KE-PE;

sol=First[NDSolve[{D[L,theta[t]]-D[L,theta'[t],t]==0,
                    theta[0]==0,
                    theta'[0]==7},{theta},{t,0,tmax=2}]]
```

```
ParametricPlot[{{xc[t],yc[t]}/.sol, {x[t],y[t]}/.sol},{t,0,tmax},PlotRange->All]
```

```
frame[t_]:=Show[Graphics[{Circle[{xc[t],yc[t]}/.sol, R],
                        Green,
                        Disk[{x[t],y[t]}/.sol,R/10],
                        Black,
                        Line[{{10 Cos[alpha],-10 Sin[alpha]},{-10 Cos[alpha],10 Sin[alpha]}}]},
                        PlotRange->10]]
```
